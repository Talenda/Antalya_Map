### Markers shows Png on charts

```python
len(north)
```

```python
for i in range(len(north)):
    names = name.iloc[i]
    elv = elevation.iloc[i]
    are = area.iloc[i]
    marker = folium.Marker(
       location=[north.iloc[i],east.iloc[i]],
       tooltip= name.iloc[i],
       icon=folium.Icon(icon="cloud", color ='darkblue'),
       popup=f"<div><p>Staion Name:{names}</p></div><b><img src=vis/vis{i}.png></b><div><p>Elevation:{elv}</p><div><div><p>Area:{are}</p><div>"
)
    marker.add_to(my_map)
```

_F string formating for markers_

```python
for i in range(len(north)):
    marker = folium.Marker(
    location=[north.iloc[i],east.iloc[i]],
       tooltip= name.iloc[i],
       icon=folium.Icon(icon="cloud", color ='darkpurple'),
       popup="<div><p>Staion Name:{}</p><div><b><img src=https://raw.githubusercontent.com/Talenda/Antalya_Map/main/Antalya_HTML/vis/vis{}.png><div><p>Elevation:{}</p><div></p><div><div><p>Area:{}</p><div></b> ".format(name.iloc[i],i,elevation.iloc[i],area.iloc[i])
)
    marker.add_to(my_map)
```
